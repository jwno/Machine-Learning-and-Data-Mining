```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import mltools as ml
%matplotlib inline
```

```python
path_to_file = 'data/curve80.txt'
data = np.genfromtxt(path_to_file, delimiter=None) # Loading the text data file
```

```python
X = data[:,0]
X = np.atleast_2d(X).T # code expects shape (M,N) so make sure it's 2-dimensional
Y = data[:,1]
Xtr,Xte,Ytr,Yte = ml.splitData(X,Y,0.75) # split data set 75/25
```

    Part 3

    (a)

```python
Xtr2 = np.zeros( (Xtr.shape[0],2) ) # create Mx2 array to store features
Xtr2[:,0] = Xtr[:,0] # place original "x" feature as X1
Xtr2[:,1] = Xtr[:,0]**2 # place "x^2" feature as X2
# Now, Xtr2 has two features about each data point: "x" and "x^2"
```

```python
def MSE(Yte, YteHat):
    return np.mean(Yte-YteHat)**2
```

```python
# Create polynomial features up to "degree"; don't create constant feature
# (the linear regression learner will add the constant feature automatically)
degrees = np.array([1, 3, 5, 7, 10, 18])
mse_error = np.zeros(degrees.shape[0])

xs = np.linspace(0, 4.2, 200)
xs = np.atleast_2d(xs).T
for i, degree in enumerate(degrees):
    XtrP = ml.transforms.fpoly(Xtr, degree, bias=False)

    # Rescale the data matrix so that the features have similar ranges / variance
    XtrP,params = ml.transforms.rescale(XtrP)

    # "params" returns the transformation parameters (shift & scale)

    # Then we can train the model on the scaled feature matrix:
    lr = ml.linear.linearRegress( XtrP, Ytr ) # create and train model

    # Now, apply the same polynomial expansion & scaling transformation to Xtest:
    XteP,_ = ml.transforms.rescale( ml.transforms.fpoly(Xte,degree,False), params)
    
    xsP = ml.transforms.fpoly(xs, degree, False)
    xsP,_ = ml.transforms.rescale(xsP, params)
    ys = lr.predict(xsP)
    
    print("Degree", degree)
    fig, ax = plt.subplots(1, 1, figsize=(10, 8)) # Create axes for single subplot
    ax.plot(xs, ys, lw=3, color='black', alpha=0.75, label='Prediction')
    ax.scatter(Xtr, Ytr, color='blue', label='Train')
    ax.scatter(Xte, Yte, color='red', label='Test')
    
    ax.set_xlim(-0.2, 4.4)
    ax.set_ylim(-3, 2) # Set the minimum and maximum limits
    ax.set_xticklabels(ax.get_xticks(), fontsize=25)
    ax.set_yticklabels(ax.get_yticks(), fontsize=25)
    
    ax.legend(fontsize=20, loc=0)
    plt.show()
    
    # MSE for Part (b)
    YteHat = lr.predict(XteP)
    mse_error[i] = MSE(Yte, YteHat)
```

    (b)

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.semilogy(degrees, mse_error, lw=4, marker='d', markersize=20, alpha=0.75, label='MSE ERROR')

ax.set_xticklabels(ax.get_xticks(), fontsize=25)
ax.set_yticklabels(ax.get_yticks(), fontsize=25)

ax.legend(fontsize=20, loc=0)
plt.show()
```

    Part 4

```python
J = np.zeros(8)
Fold_Array = np.array([2,3,4,5,6,10,12,15]);
for i, nFolds in enumerate(Fold_Array):
    K = np.zeros(nFolds)
    for iFold in range(nFolds):
        Xti,Xvi,Yti,Yvi = ml.crossValidate(Xtr,Ytr,nFolds,iFold) # use ith block as validation
        XtiP = ml.transforms.fpoly(Xti,5,False)
        learner = ml.linear.linearRegress(XtiP, Yti) # TODO: train on Xti, Yti, the data for this fold

        XviP = ml.transforms.fpoly(Xvi,5,False)
        XviP = learner.predict(XviP)
        K[iFold] = MSE(Yvi, XviP) # TODO: now compute the MSE on Xvi, Yvi and save it
    J[i] = np.mean(K)
# the overall estimated validation error is the average of the error on each fold
print(np.mean(J))

fig, ax = plt.subplots(1, 1, figsize=(10, 8))

ax.semilogy(Fold_Array, J, lw=4, marker='d', markersize=20, alpha=0.75, label='MSE ERROR')

ax.set_xlim(0.7, 16)
ax.set_ylim(0.01, 0.25)

ax.set_xticklabels(ax.get_xticks(), fontsize=25)
ax.set_yticklabels(ax.get_yticks(), fontsize=25)

ax.legend(fontsize=20, loc=0)
plt.show()
```

As the number of folds increase, the MSE error increases significantly at 4, 
but becomes bit steady. However, the rate of error is not as variant as it 
ranges from 0.01 to 0.25 depending on the fold.
